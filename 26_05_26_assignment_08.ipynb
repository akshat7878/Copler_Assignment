Also Explain the primary differences between df.join() and pd.merge(),

```python
dfA = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya']
})

dfB = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Karan', 'Neha']
})

dfC = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'City': ['Jaipur', 'Delhi', 'Mumbai', 'Pune']
})

print(dfA)
print('='*20)
print(dfB)
print('='*20)
print(dfC)
```

```python
vertical = pd.concat([dfA, dfB], axis=0)
vertical
```

```python
final = pd.merge(vertical, dfC, on='ID')
final
```

### Difference Between `join()` and `merge()`

| `df.join()` | `pd.merge()` |
|---|---|
| Joins mainly on index | Joins on columns or index |
| Simpler syntax | More powerful and flexible |
| Default join is left | Default join is inner |
| Best for index-based operations | Best for SQL-style operations |
| Used with DataFrames | Can merge Series and DataFrames |

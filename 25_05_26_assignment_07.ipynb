Pandas Series

```python
import pandas as pd

myDict = {
    'name': ['a', 'b', 'c', 'd'],
    'class': ['EEE', 'ECE', 'CSE', 'CE']
}

s1 = pd.Series(myDict)
print(s1)
print(type(s1))
```

```python
L = [0, 1, 2, 3, 4, 5]

s2 = pd.Series(L)
print(s2)
print(type(s2))
```

```python
#Accessing
print(s1['name'])
print('-'*30)
print(s2[0:2])
```

Pandas DataFrame

```python
cols = ['a', 'b', 'c', 'd']
List = [
    [1, 2, 3, 4],
    [4, 3, 2, 1],
    [5, 6, 7, 8],
    [9, 8, 7, 6]
]

df = pd.DataFrame(List, columns=cols)
df
```

```python
Dict = {
        'Subject': ['Maths', 'Physics', 'Chemistry'],
        'Marks'  : [60, 50, 57]
        }

df = pd.DataFrame(Dict)

df['Percentage'] = round(df['Marks'] * 100 / 70, 2)

df
```

```python
cols = ('employee', 'salary')
tup = [
    ('akshat', '4000'),
    ('vidhi', '5000'),
    ('aman', '6000')
]

df = pd.DataFrame(tup, columns=cols)
df
```

```python
list_of_dict = [
    {"Name": "Akshat", "Marks": 80},
    {"Name": "Rahul", "Marks": 90},
    {"Name": "Aman", "Marks": 75}
]

df = pd.DataFrame(list_of_dict)
df
```

Data Iteration

```python
data = {
    'Name': ['Akshat', 'Rahul', 'Aman', 'Riya'],
    'Marks': [80, 90, 75, 60],
    'City': ['Jaipur', 'Delhi', 'Mumbai', 'Jaipur']
}

df = pd.DataFrame(data)
df
```

```python
print(df[df['Marks'] > 70])
print(f"\n{'-'*30}\n")
print(df[df['City'] == 'Jaipur'])
```

```python
print(df.iloc[0])
print(f"\n{'-'*30}\n")
print(df.iloc[2:4])
```

```python
print(df.loc[0:1, 'Name'])
print(f"\n{'-'*30}\n")
print(df.iloc[0:2, 0:2])
```

```python
res = df[df['Marks'] >= 70]

print(res)
```
